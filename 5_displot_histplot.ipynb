```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/refs/heads/main/merc.csv')
df = df[df.year >= 2010]
df = df[df['transmission'] != 'Other']
df = df[df['fuelType'] != 'Other']
df.head()
```

```python
df.shape
```

```python
plt.figure(figsize=(10,5))
sns.histplot(df.price) #normal taqsimot
plt.show()
```

```python
# plt.figure(figsize=(10,5)) #ishlamaydi
sns.displot(data = df, x = df.price,  hue= 'transmission', height=5, aspect = 2 ) #normal taqsimot
plt.show()
```

```python
sns.displot(data = df, x = df.price,  col= 'transmission', height=5, aspect = 2 ) #normal taqsimot
plt.show()
```

```python
sns.displot(data = df, x = df.mileage,  hue= 'transmission', multiple = 'stack', height=5, aspect = 2 ) #normal taqsimot
plt.show()
```

```python
sns.displot(data = df, x = df.price, kind = 'kde',   height=5, aspect = 2 ) #zichligini ko'rsatish
plt.show()
```

```python
sns.displot(data = df, x = df.price, kde = True,  height=5, aspect = 2 ) #zichligini ko'rsatish va distibrution
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.kdeplot(data = df, x = df.price, hue = 'fuelType', shade = True)
plt.show()
```

```python
sns.displot(data = df, x = df.price, kind = 'kde',  hue = 'fuelType',  height=5, aspect = 2 ) #zichligini ko'rsatish
plt.show()
```

```python
sns.displot(data = df, x = df.price, kind = 'ecdf',   height=5, aspect = 2 )
plt.xticks(rotation = 90)
plt.grid(which = 'both')
plt.minorticks_on()
plt.show() #misol: mashinlarning 50%i,  20000 dan pastda
```

```python
sns.displot(data = df, x = df.mileage, kind = 'ecdf',   height=5, aspect = 2 )
plt.xticks(rotation = 90)
plt.grid(which = 'both')
plt.minorticks_on()
plt.show() #misol: mashinlarning 50%i,  25000 mileage dan kam yurgan
```
