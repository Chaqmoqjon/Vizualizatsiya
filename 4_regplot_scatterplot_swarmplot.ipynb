```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/refs/heads/main/merc.csv')
df.head()
```

```python
df = df[df['year']  > 1990] #ortiqcha bitta yil o'chirildi
```

```python
plt.figure(figsize = (10, 6))
# sns.scatterplot(x = 'year', y = 'price', data = df)
sns.regplot(x = 'year', y = 'price', data = df) #o'suvchi ijobiy korrelatsiya bor ekan
plt.show()
```

```python
plt.figure(figsize = (10, 6))
sns.scatterplot(x = 'mileage', y = 'price', data = df) #mashinalarning yurgan yo'li oshgan sari, narxi tushib bormoqda
# sns.regplot(x = 'mileage', y = 'price', data = df) #o'suvchi salbiy korrelatsiya bor ekan
plt.show()
```

```python
plt.figure(figsize = (18, 8))
sns.scatterplot( data = df, x = 'mileage', y = 'price', hue = 'transmission', style = 'fuelType', s=100) #hue bilan har bir guruhga alohida rang berildi
# sns.regplot(x = 'mileage', y = 'price', data = df) #o'suvchi salbiy korrelatsiya bor ekan
plt.show()
```

```python
plt.figure(figsize = (16, 6))
sns.scatterplot( data = df, x = 'mileage', y = 'price', hue = 'fuelType', s=100) #petrol dyuradigan mashinalar eng qimmatlari
plt.show()
```

```python
sns.swarmplot(data = df, x = 'fuelType', y = 'price') # Ishlatmaganing yaxshi! tahminan 20 daqiqa vaqt ketti
plt.show()
```
