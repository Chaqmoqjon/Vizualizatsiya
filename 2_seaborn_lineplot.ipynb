```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/refs/heads/main/museum_visitors.csv', index_col=0)
df.head()
```

```python
df.info()
```

```python
sns.lineplot(data=df) #ko'proq ma'lumotlar bilan
plt.show()
```

```python
plt.plot(df) #kamroq ma'lumotlar bilan
plt.show()
```

```python
plt.figure(figsize=(15, 5)) #plt va sns birlashmasi
plt.title('Muzeydagi tashriflar soni')
plt.xlabel('Yil')
plt.ylabel('Tashriflar soni')
plt.xticks(rotation=90)
plt.grid()
sns.lineplot(data=df)
plt.show()
```
