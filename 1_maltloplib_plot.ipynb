```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
sonlar = np.arange(10)
sonlar
```

```python
plt.plot(sonlar) #chizmali garfik
plt.show()
```

```python
x = np.arange(0, 4*np.pi, 0.1)
y = np.sin(x)
```

```python
plt.figure(figsize = (10, 5))
plt.plot(x, y, linewidth = 2, linestyle = "-", color = "r", marker = "o")
plt.title("Sinus")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(color = 'b', linestyle = '--')
plt.show()
```

```python
x = np.arange(0, 4*np.pi, 0.1)
sin = np.sin(x)
cos = np.cos(x)
plt.figure(figsize = (10, 5))

plt.plot(x, sin, linewidth = 2, linestyle = "-", color = "r", alpha = 0.7,   marker = "h")
plt.plot(x, cos, linewidth = 2, linestyle = "-", color = "b", alpha = 0.7,  marker = "v")

plt.title("Sinus va Cosinus funksiya grafigi")
plt.xlabel("x")
plt.legend(['sin(x)', 'cos(x)'], loc = 0, fontsize = 'x-large')
plt.grid(color = 'tomato', alpha = 0.5,  linestyle = '--')
plt.savefig('grafik.png')
plt.show()
```
