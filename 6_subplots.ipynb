```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = np.arange(0, 4 * np.pi, 0.1)
sin = np.sin(x) # sin(x)
cos = np.cos(x) # cos(x)

plt.figure(figsize=(12,6))
plt.subplot(2,1,1) # 2 - jami qatorlar soni, 1 - jami ustunlar soni, 1-grafik
# sin(x) grafigi
plt.plot(x, sin,
         linewidth=1.5,
         linestyle='-.',
         color='#8E44AD',
         alpha=0.7,
         marker='h')
plt.title('sin(x) funksiya grafigi')
plt.grid(color='tomato', alpha=0.5, linestyle='--') # setkani o'zgartirish

plt.subplot(2,1,2) # # 2 - jami qatorlar soni, 1 - jami ustunlar soni, 2-grafik
# cos(x) grafigi
plt.plot(x, cos,
         linewidth=1.5,
         linestyle=':',
         color='c',
         alpha=0.7,
         marker='v')
plt.title('cos(x) funksiya grafigi')
plt.grid(color='tomato', alpha=0.5, linestyle='--') # setkani o'zgartirish

plt.show()
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/refs/heads/main/merc.csv')
df = df[df.year >= 2010]
df = df[df['transmission'] != 'Other']
df = df[df['fuelType'] != 'Other']
df.head()
```

```python
fig1 = plt.figure(figsize= (12, 6))
```

```python
sns.histplot(data = df, x = 'price')
plt.show()
```

```python
sns.scatterplot(data = df, x = 'price', y= 'year', s = 100)
plt.show()
```

```python
fig, ax = plt.subplots(2, 1, figsize = (10, 10)) #ikki vaundan ortiq grafik chizish
sns.histplot(ax = ax[0], data = df, x = 'price')
sns.scatterplot(ax = ax[1], data = df, x = 'price', y= 'year', s = 100)
```

```python
fig, ax = plt.subplots(2, 2, figsize = (15, 10))
fig.suptitle('Car price analysis', fontsize = 20)
sns.histplot(ax = ax[0, 0], data = df, x = 'price')
sns.histplot(ax = ax[0, 1], data = df, x = 'year')
sns.scatterplot(ax = ax[1, 0], data = df, x = 'year', y= 'mileage', hue = 'fuelType', s = 50)
sns.scatterplot(ax = ax[1, 1], data = df, x = 'year', y = 'price', hue = 'fuelType', s = 50)
ax[0, 0].set_title('Price distribution')
ax[0, 1].set_title('Year distribution')
ax[1, 0].set_title('Mileage vs Year')
ax[1, 1].set_title('Price vs Year')
plt.show()
```
